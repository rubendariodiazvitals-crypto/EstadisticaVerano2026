```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/rubendariodiazvitals-crypto/EstadisticaVerano2026/refs/heads/main/StudentPerformance/Student_Performance.csv")

df
```

Establecemos las horas de estudio (Hours_Studied) como variable dependiente y el indice de desempeño ("Perfomance_Index") como la variable independiente.Esto es porque la hipotesis inicial es que cambios en las horas de estudio puede hacer cambios en el indice de desempeño no causa cambios en las horas de estudio

```python
X = df["Hours_Studied"]       # Variable independiente
Y = df["Performance_Index"]     # Variable dependiente
```

2. **Realice un diagrama de dispercion**

```python
import matplotlib.pyplot as plt

# Configuración del gráfico
plt.figure(figsize=(6, 4), dpi=120)

# Gráfico de dispersión con tus datos reales
plt.scatter(
    X, Y,
    marker="o",
    color='blue',
    edgecolor='black',
    alpha=0.8,
    s=50
)

# Etiquetas personalizadas para tu contexto
plt.xlabel('ausencias', fontsize=12)
plt.ylabel('salarios', fontsize=12)

# Ajustes visuales
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.grid(visible=True, linestyle='--', linewidth=0.7, alpha=0.5, color="gray")

# Mostrar el gráfico
plt.show()
```

Una vez ajustado el modelo de regresion lineal

El grafico de dispersion muestra una clara tendencia lineal. Se puede observar que conforme aumentan las horas de estudio (X) aumenta el indice de desempeño (Y).Con lo que se espera un coeficiente de correlacion positivo y parece cumplirse el supuesto de lineallidad.

```python
# Importar la función pearsonr desde scipy.stats
from scipy.stats import pearsonr

# Test de Pearson
# H0: rho = 0     (No hay correlación)
# H1: rho ≠ 0     (Sí hay correlación)
# alpha = 0.05

r, valor_p = pearsonr(X, Y)

print(f"Coeficiente de correlación: {r: 0.4f}")
print(f'valor_p: {valor_p: 0.4f}')
```

El coeficiente de correlacion $r=0.3737$
